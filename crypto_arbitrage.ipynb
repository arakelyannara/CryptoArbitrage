#### 2. For each of the three dates, calculate the spread returns.
```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

pos_val=arbitrage_spread_early>0
arbitrage_spread_early=arbitrage_spread_early[pos_val]
spread_return_early= arbitrage_spread_early/coinbase_sliced.loc[early_date]
spread_return_early=spread_return_early.dropna()
# Review the spread return DataFrame
spread_return_early.head()
```

```python

pos_val=arbitrage_spread_middle>0
arbitrage_spread_middle=arbitrage_spread_middle[pos_val]
spread_return_middle= arbitrage_spread_middle/coinbase_sliced.loc[middle_date]
spread_return_middle=spread_return_middle.dropna()

# Review the spread return DataFrame
spread_return_middle.head()
```
